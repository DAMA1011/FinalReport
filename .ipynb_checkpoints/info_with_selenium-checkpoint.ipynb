```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
from urllib.parse import quote
import re
import requests
import sys
import os
import time
import json
import html2text
import pandas as pd
import undetected_chromedriver as uc
from datetime import datetime
import json
```

```python
def initialize_chrome(_from="facebook",retry=0):
    global driver 
    try:
        print("Initializing chromedriver.")
        options = Options()
        options.add_argument("--start-maximized")
        options.add_argument("--incognito")
        options.add_argument("--disable-gpu")
        options.add_argument("--headless")
        options.add_argument('--log-level=3')
        driver = uc.Chrome(options=options)
        time.sleep(2)
        return True
    except Exception as e:
        print(e)
        pass

def page_scroll():
    time.sleep(2)
    while True:
        windows = driver.find_element(By.XPATH, '//*[@id="QA0Szd"]/div/div/div[1]/div[2]/div/div[1]/div/div/div[2]/div[1]')
        try:        
            end = WebDriverWait(driver, 2).until(EC.presence_of_element_located( (By.CSS_SELECTOR, "span.HlvSq") ))
            if end:
                print('到底了')
                break
        except:
            driver.refresh()
            windows = driver.find_element(By.XPATH, '//*[@id="QA0Szd"]/div/div/div[1]/div[2]/div/div[1]/div/div/div[2]/div[1]')
            time.sleep(0.2)
            count = 0
            limit = 10
            while count <= limit:
                innerHeight_1 = driver.execute_script('return arguments[0].scrollHeight;', windows)
                driver.find_element(By.XPATH, '//*[@id="QA0Szd"]/div/div/div[1]/div[2]/div/div[1]/div/div/div[2]/div[1]').send_keys(Keys.PAGE_DOWN)
                time.sleep(0.2)
                driver.find_element(By.XPATH, '//*[@id="QA0Szd"]/div/div/div[1]/div[2]/div/div[1]/div/div/div[2]/div[1]').send_keys(Keys.PAGE_DOWN)
                time.sleep(0.2)
                driver.find_element(By.XPATH, '//*[@id="QA0Szd"]/div/div/div[1]/div[2]/div/div[1]/div/div/div[2]/div[1]').send_keys(Keys.PAGE_DOWN) 
                time.sleep(0.2)
                driver.find_element(By.XPATH, '//*[@id="QA0Szd"]/div/div/div[1]/div[2]/div/div[1]/div/div/div[2]/div[1]').send_keys(Keys.PAGE_DOWN)
                time.sleep(0.2)
                innerHeight_2 = driver.execute_script('return arguments[0].scrollHeight;', windows)
                if innerHeight_1 == innerHeight_2:
                    count += 1
            print('捲不動了')
            
def get_url():
    url_data = []
    response = driver.page_source
    soup = BeautifulSoup(response, 'html.parser')
    url = soup.select("div[data-js-log-root] a[aria-label]")
    for item in url:
        url_dict = {}
        url_dict['href'] = item.get('href')
        url_data.append(url_dict)
    df = pd.DataFrame(url_data)
    df.to_csv('test.csv', mode = 'a', header=False)
```

```python
time1 = time.time()
district = "台北市大同區"
street = ["五原路", "伊寧街", "保安街", "南京西路", "哈密街", "塔城街"]
category = ["火鍋", "拉麵", "日式", "美式"]
initialize_chrome()
for item_a in street:
    for item_b in category:
        keywords = f"{district}{item_a}{item_b}"
        url = f"https://www.google.com.tw/maps/search/{keywords}/@25.0303169,121.5029609,14z/data=!3m1!4b1"
        driver.get(url)
        time.sleep(3)
        page_scroll()
        get_url()
        print(keywords + " 完成")
driver.quit()
print(f'執行總花費時間: {time.time() - time1}')
```
