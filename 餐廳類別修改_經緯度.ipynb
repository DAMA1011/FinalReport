## 方法1 用dict更改

```python
# 把原本的類別縮減成15大類別
Series_ = dfFilter1["place_category"].map(dfFilter2.to_dict())
# # 把餐廳類別賦予15大類別
dict2 = dfFilter3.loc[:,"標記"].to_dict()
dict1 = Series_.to_dict()
dict1.update(dict2)
# # 把其他類別更改成15大類別
dict3 = dfFilter4.loc[:,"標記"].to_dict()
dict1.update(dict3)
pd.Series(dict1)
```

## 方法2 直接用Series更改


```python
# 把原本的類別縮減成15大類別
Series1 = dfFilter1["place_category"].map(dfFilter2.to_dict())
# 把餐廳類別賦予15大類別
Series2 = dfFilter3.loc[:,"標記"]
Series1.update(Series2)
# 把其他類別更改成15大類別
Series3 = dfFilter4.loc[:,"標記"]
Series1.update(Series3)
Series1
```
